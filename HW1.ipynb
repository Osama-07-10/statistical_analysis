Sex vs Survived:
Females had a much higher survival rate than males, indicating the â€œwomen and children firstâ€ evacuation policy.

Pclass vs Survived:
Passengers in 1st class survived significantly more compared to 2nd and 3rd class. Higher socioeconomic status increased survival chances.


Age vs Survived:
Younger passengers, especially children, showed higher survival rates, while survival decreased with increasing age.


Fare vs Survived:
Higher ticket fares were associated with higher survival. This supports the idea that passengers who paid more (usually 1st class) were closer to lifeboats.
